```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("dataset/customer_shopping_behavior.csv")
df
```

```python
df.columns=df.columns.map(lambda x: (x.replace(' ','_')).lower())
```

```python
df.columns
```

```python
(df.discount_applied==df.promo_code_used).all()
```

```python
df.drop(columns=['promo_code_used'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['review_rating']=df['review_rating'].fillna(df.groupby('category')['review_rating'].transform('median'))
```

```python
df.isnull().sum()
```

```python
df.frequency_of_purchases.unique()
```

```python
word_to_days = {
    'Weekly'        : 7,
    'Fortnightly'   : 14,
    'Bi-Weekly'     : 14,
    'Monthly'       : 30,
    'Every 3 Months': 90,
    'Quarterly'     : 90,
    'Annually'      : 365
}
df.frequency_of_purchases=df.frequency_of_purchases.map(word_to_days)
```

```python
df
```

```python
df['Age_Group']=pd.qcut(df['age'], q=4, labels=['Young Adults', 'Adults', 'Middle Aged', 'Senior Citizen'])
```
